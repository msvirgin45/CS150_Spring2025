### Task 2: Computing Basic Statistics

##### I choose to focus on the categorical column of a player's position (pos) and on two continous numerical columns of points (pts) and assists (ast).

##### Below are the summary statistics: mean, max, min, std dev, range, and median

#### Compute the mean, maximum, minimum, standard deviation, range, and median for the two numeric columns that you chose.

```python
print("Overall Points (pts) Stats:")
print("Mean:", df['pts'].mean())
print("Max:", df['pts'].max())
print("Min:", df['pts'].min())
print("Standard Deviation:", df['pts'].std())
print("Median:", df['pts'].median())

print("\nOverall Assists (ast) Stats:")
print("Mean:", df['ast'].mean())
print("Max:", df['ast'].max())
print("Min:", df['ast'].min())
print("Standard Deviation:", df['ast'].std())
print("Median:", df['ast'].median())
```

####  Compute the mean, maximum, minimum, standard deviation, range, and median for the two numeric columns that you chose across each available value in the category column.

```python
grouped = df.groupby('pos')[['pts', 'ast']] #Code that does this task for all labeled positions.

print('Positional mean:')
print(grouped.mean())
print()
print('Positional maximum:')
print(grouped.max())
print()
print('Positional minimum:')
print(grouped.min())
print()
print('Positional standard deviation:') #NaN means when calculating there was an error (ex: 0 being divided by)
print(grouped.std())
print()
print('Positional median:')
print(grouped.median())
```

```python
#Filter for only common positions
common_positions = ['PG', 'SG', 'SF', 'PF', 'C']
df_common = df[df['pos'].isin(common_positions)]

#Group each position together and calculate statistics for pts and ast
grouped_common = df_common.groupby('pos')[['pts', 'ast']]

print("Mean for Points and Assists by Position:")
print(grouped_common.mean())
print()
print("Maximum for Points and Assists by Position:")
print(grouped_common.max())
print()
print("Minimum for Points and Assists by Position:")
print(grouped_common.min())
print()
print("Standard Deviation for Points and Assists by Position:")
print(grouped_common.std())
print()
print("Median for Points and Assists by Position:")
print(grouped_common.median())
```

### Task 3: Plotting Distributions

##### Here are plots showing relationships between the numerical columns and category values.

#### Plot a scatter plot depicting the two continuous variables against each other one should be on the x axis, and the other on the y-axis.

```python
import matplotlib.pyplot as plt

df.plot.scatter(x='pts', y='ast', title='Points vs Assists')
plt.show()
```

#### Plot a histogram (bar plot) for each of each of your continuous variables on separate plots.

```python
# Histogram for Points
df['pts'].plot.hist(bins=30, title='Points Distribution')
plt.xlabel('Points')
plt.show()

# Histogram for Assists
df['ast'].plot.hist(bins=30, title='Assists Distribution')
plt.xlabel('Assists')
plt.show()
```

#### Choose one of your continuous measures and repeat (ii), but plot the histograms for each category for your category column.

```python
#Filters for common positions (PG, SG, SF, PF, C)
positions = ['PG', 'SG', 'SF', 'PF', 'C']
df_filtered = df[df['pos'].isin(positions)]

#Makes seperate histograms for each posistion
for pos in positions:
    df_pos = df_filtered[df_filtered['pos'] == pos]
    df_pos['pts'].plot.hist(bins=30, title=f'Points Distribution for {pos}')
    plt.xlabel('Points')
    plt.show()
```
